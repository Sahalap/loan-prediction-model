```python
df.drop(['loan_id','education','self_employed','commercial_assets_value','bank_asset_value'],inplace=True,axis=1)
```

```python
df.head()
```

```python
x = df.drop('loan_status',axis=1)
```

```python
y = df['loan_status']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test = train_test_split(x,y, test_size=0.25,random_state=4)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
import joblib

joblib.dump(model, 'logistic_regression.pkl')
```

```python
lr_model = joblib.load('logistic_regression.pkl')
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()

model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
import joblib

joblib.dump(model, 'Decision_tree.pkl')
```

```python
dt_model = joblib.load('Decision_tree.pkl')
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()

model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
